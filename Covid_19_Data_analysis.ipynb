```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
covid_deaths = pd.read_csv("/content/drive/MyDrive/Covid /CovidDeaths (3).csv")
covid_vaccination = pd.read_csv("/content/drive/MyDrive/Covid /CovidVaccinations.c")
```

```python
"""
select location , date , total_cases , new_cases , total_deaths, population
from covid_deaths
order by 1,2
"""


selected_columns = ['location', 'date', 'total_cases', 'new_cases', 'total_deaths', 'population']
covid_deaths_sorted = covid_deaths.sort_values(by=[ "location"],ascending=[True])[selected_columns]


covid_deaths_sorted
```

```python
# Total Cases in a country vs total deaths

"""
select location , date , total_cases , total_deaths , (total_deaths/total_cases)*100 AS Percentage_Death
from covid_deaths ;
"""
covid_deaths_result = covid_deaths[['location', 'date', 'total_cases', 'total_deaths']]
covid_deaths_result['Percentage_Death'] = (covid_deaths['total_deaths'] / covid_deaths['total_cases']) * 100


covid_deaths_result
```

```python

covid_deaths_india = covid_deaths[covid_deaths['location'] == 'India']

covid_deaths_result = covid_deaths_india[['location', 'date', 'total_cases', 'total_deaths']]
covid_deaths_result['Percentage_Death'] = (covid_deaths_india['total_deaths'] / covid_deaths_india['total_cases']) * 100





covid_deaths_result
```

```python
"""
Select total_deaths, Population, Max((total_cases/population))*100 as PercentPopulationInfected
From PortfolioProject..CovidDeaths
--Where location like '%states%'
Group by Location, Population
order by PercentPopulationInfected desc
"""



grouped_data = covid_deaths.groupby(['location'])['total_cases'].max()
group_data_population = covid_deaths.groupby(['location'])['population'].max()




percentage_infected = (grouped_data / group_data_population) * 100


result_df = pd.concat([grouped_data,group_data_population, percentage_infected], axis=1)
result_df.columns = ['Highest_case_in_a_Day',"Population", 'PercentPopulationInfected']
result_sorted = result_df.sort_values(by = ["PercentPopulationInfected"], ascending = [False])


result_sorted
```

```python
"""
Select continent, sum(Total_deaths as int) as TotalDeathCount
From PortfolioProject..CovidDeaths
--Where location like '%states%'
Where continent is not null
Group by continent
order by TotalDeathCount desc
"""

filtered_data = covid_deaths[covid_deaths['continent'].notnull()]
grouped_data = filtered_data.groupby('continent')['total_deaths'].sum().reset_index()
grouped_data.rename(columns={'Total_deaths': 'TotalDeathCount'}, inplace=True)

sorted_data = grouped_data.sort_values(by = ["total_deaths"],ascending = [False])
sorted_data


```

```python
plt.figure(figsize=(10, 5))
plt.bar(sorted_data['continent'], sorted_data['total_deaths'])
plt.xlabel('Continent')
plt.ylabel('Total Death Count')
plt.title('Total Death Count by Continent')
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()


```

```python
"""
Select dea.continent, dea.location, dea.date, dea.population, vac.new_vaccinations
From PortfolioProject..CovidDeaths dea
Join PortfolioProject..CovidVaccinations vac
	On dea.location = vac.location
	and dea.date = vac.date
where dea.continent is not null
order by 2,3
"""

filtered_data_deaths = covid_deaths[covid_deaths['continent'].notnull()]

# Join covid_deaths and covid_vaccinations on 'location' and 'date'
merged_data = pd.merge(filtered_data_deaths, covid_vaccination, on=['location', 'date'], how='inner')



# Select the required columns for the final result
result_df = merged_data[['continent_x', 'location', 'date', 'population', 'new_vaccinations']]

# Display the result DataFrame
result_df
```
